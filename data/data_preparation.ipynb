The dataset we are dealing with was obtained from Kaggle at https://www.kaggle.com/datasets/shivamb/netflix-shows. 
This dataset consists of 8807 rows and 12 features:

- show_id: Netflix ID of the media.
- type: Movie or TV Show.
- title: Title of the media.
- director: Director of the media.
- cast: Cast of the media.
- country: Country in which the movie was made.
- date_added: Date in which the media was added.
- release_year: Year in which the media was released.
- rating: Age rating of the media.
- duration: Duration of the media.
- listen_in: Classification given by Netflix.
- description: A short description of the media.

# Missing Data

```python
netflix.isnull().sum()
```

```python
#Show missing values in %

header = ["Column", "Missing Value (%)"]
missing_value_data = []
for col in netflix.columns:
    nulls = (netflix[col].isna().sum() / len(netflix) * 100)
    if nulls > 0 :
        missing_value_data.append([col, round(nulls,2)])
        #print("{} missing values: {}%".format(col, round(nulls,2)))
        
print(tabulate(missing_value_data, headers=header, tablefmt="fancy_grid"))
```

### Dealing with missing values

There are 6 features with missing values.
We must considers first three rows of the previous table, since they have more than 9% missing values. The others can be dropped.

"Director" and "Cast" will be replaced with "Data not available".

```python
netflix['director'].replace(np.nan, 'Data not available', inplace = True)
netflix['cast'].replace(np.nan, 'Data not available', inplace = True)
```

```python
netflix[netflix['country'].isnull()]
```

I will replace "country" missing values with the most common country (instead of dropping all of them)

```python
netflix['country'] = netflix['country'].fillna(netflix['country'].mode()[0])
```

```python
# Drops all the others
netflix.dropna(inplace = True)
```

```python
netflix.isnull().sum()
```

# Duplicate

```python
# Drop Duplicates
netflix.drop_duplicates(inplace= True)
```

# Data Transformation

#### "date_added"

```python
netflix["date_added"] = pd.to_datetime(netflix['date_added'])
```

```python
netflix['month_added']=netflix['date_added'].dt.month
netflix['month_name_added']=netflix['date_added'].dt.month_name()
netflix['year_added'] = netflix['date_added'].dt.year
```

```python
netflix.head(3)
```

#### "listed_in"

```python
netflix["listed_in"].unique()
```

I'll take only the first 3 categories.

```python
netflix.shape[0]
```

```python
netflix['listed_in_1'] = 0
netflix['listed_in_2'] = 0
netflix['listed_in_3'] = 0

tmp = netflix['listed_in'].str.split(',')

for i, row in netflix.iterrows():
    t = len(tmp[i])
    if t == 1:
        netflix['listed_in_1'][i] = tmp[i][0]
        netflix['listed_in_2'][i] = 0
        netflix['listed_in_3'][i] = 0
    if t == 2:
        netflix['listed_in_1'][i] = tmp[i][0]
        netflix['listed_in_2'][i] = tmp[i][1]
        netflix['listed_in_3'][i] = 0
    if t == 3:
        netflix['listed_in_1'][i] = tmp[i][0]
        netflix['listed_in_2'][i] = tmp[i][1]
        netflix['listed_in_3'][i] = tmp[i][2]
```
